ALGORITHM : <br>
1. Start the program.
2. Create a function and initialize the variables.
3. Take the weights, input values and make a dot product of them.
4. From the dot product pass the result to the signum function.
5. Check with the calculated with the target value.
6. If target differs, calculate the weights.
7. Continue the Epochs until all the target match with the calculated value.
8. Print the final weights and stop the program.

ANN LOGIC GATES <br>
External libraries not used 😎


```python
def ANN_LOGIC_GATE(Input_x1, Input_x2, Output):
    # Initializing variables
    w1, w2, b = 0, 0, 0
    Epochs = len(Output)
    Learning_Rate = 1
    Yes_No = ""

    # Signum function
    def calc(eqn):
        Calc_Val_Return = None

        if eqn > 0:
            Calc_Val_Return = 1
        elif eqn < 0:
            Calc_Val_Return = -1
        elif eqn == 0:
            Calc_Val_Return = 0

        return Calc_Val_Return

    # Epochs
    for epoch in range(Epochs):

        # Count to check on the target and calc match
        Yes_Count = 0

        print(f"\nEPOCH : {epoch+1}\n")

        print(
            "x1 ### x2 ### Tar ### Yin ### Calc ### del w1 ### del w2 ### del b ### w1 ### w2 ### b"
        )

        for i in range(len(Input_x1)):
            Y_in = b + (Input_x1[i] * w1) + (Input_x2[i] * w2)

            # Storing in another variable
            pre_w1 = w1
            pre_w2 = w2
            pre_b = b

            # Check if Target and calculated are same
            if calc(Y_in) == Output[i]:
                # If same, then no changes in the weights
                Yes_No = "Y"
                Yes_Count += 1

                w1, w2, b = 0, 0, 0

                Del_w1 = 0
                Del_w2 = 0
                Del_b0 = 0

            else:
                # Target not matching, Change the weights
                Yes_No = "N"

                Del_w1 = Learning_Rate * Output[i] * Input_x1[i]
                Del_w2 = Learning_Rate * Output[i] * Input_x2[i]
                Del_b0 = Learning_Rate * Output[i]

            # Update new weights
            w1 = Del_w1 + pre_w1
            w2 = Del_w2 + pre_w2
            b = Del_b0 + pre_b

            # Show the intermediate steps
            print(
                f"{Input_x1[i]} ### {Input_x2[i]} ### {Output[i]}  -- {Yes_No} -- ### {Y_in} ### {calc(Y_in)} -- {Yes_No} -- ### {Del_w1} ### {Del_w2} ### {Del_b0} ### {w1} ### {w2} ### {b}"
            )

        # Yearly stopping
        if Yes_Count == len(Output):
            print(f"\nEARLY STOPPING DONE AT EPOCH {epoch+1}\n")
            break

    # Final show
    print("\nFINAL WEIGHTS\n")
    print(f"W1 : {w1}")
    print(f"W2 : {w2}")
    print(f"b : {b}")
```

AND GATE


```python
# Initializing gates
Input_x1 = [1, 1, -1, -1]
Input_x1_t = [-1,-1,1,1]

Input_x2 = [1, -1, 1, -1]
Input_x1_t = [-1,1,-1,1]

Output = [1, -1, -1, -1]

ANN_LOGIC_GATE(Input_x1, Input_x2, Output)
print("AND Gate")
```

OR GATE


```python
# Initializing gates
Input_x1 = [1, 1, -1, -1]
Input_x2 = [1, -1, 1, -1]
Output = [1, 1, 1, -1]

ANN_LOGIC_GATE(Input_x1, Input_x2, Output)
print("OR Gate")
```

XOR Gate


```python
# Initializing gates
Input_x1 = [1, 1, -1, -1]
Input_x2 = [1, -1, 1, -1]
Output = [1, -1, -1, 1]

ANN_LOGIC_GATE(Input_x1, Input_x2, Output)
print("XOR Gate")
```
