```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#df = pd.read_csv("/Users/raghav/Documents/Programs/python/ml/datasets/ad-postive-correl.csv")
df = pd.read_csv("/Users/raghav/Documents/Programs/python/ml/datasets/cars-neg-correl.csv")
```

```python
x = np.array(df.iloc[:,0]).reshape((-1,1))
y = np.array(df.iloc[:,1])
```

```python
model = LinearRegression().fit(x,y)
```

```python
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
```

```python
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```

```python
eqn = f'{model.intercept_} = {model.coef_[0]}*x + c'
eqn
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred, color = 'red')
plt.show()
```
